```python
from quip import QuIPQuantizer
from transformers import AutoModelForCausalLM, AutoTokenizer

# Load the pre-trained model and tokenizer
model_name = "meta-llama/Llama-2-7b-hf"
model = AutoModelForCausalLM.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Initialize QuIP quantizer
quantizer = QuIPQuantizer(model)

# Perform 2-bit quantization
quantized_model = quantizer.quantize(bits=2)

# Save the quantized model
quantized_model.save_pretrained("llama2_7b_2bit")
tokenizer.save_pretrained("llama2_7b_2bit")
```
